#  Introducción a Seaborn

Seaborn es una biblioteca de visualización estadística en Python basada en `matplotlib`.

Ofrece una **API de alto nivel**, más concisa y atractiva, ideal para:
- Visualización rápida de DataFrames
- Análisis exploratorio
- Estética automática de calidad

*Seaborn trabaja por debajo con matplotlib y pandas*

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset("tips")
df.head()
```


## Funciones más usadas en Seaborn (con parámetros comunes)

| Función             | Descripción breve                                         | Ejemplo práctico                                          | Parámetros principales                                             |
|---------------------|-----------------------------------------------------------|------------------------------------------------------------|--------------------------------------------------------------------|
| `sns.scatterplot()` | Gráfico de dispersión entre dos variables numéricas.      | `sns.scatterplot(data=df, x="total_bill", y="tip")`        | `data`, `x`, `y`, `hue`, `style`, `size`, `palette`, `alpha`      |
| `sns.lineplot()`    | Gráfico de líneas (útil para series o relaciones continuas).| `sns.lineplot(data=df, x="x", y="y")`                     | `data`, `x`, `y`, `hue`, `style`, `markers`, `dashes`, `ci`       |
| `sns.barplot()`     | Barras que representan medias o estimaciones por categoría.| `sns.barplot(data=df, x="sex", y="tip")`                  | `data`, `x`, `y`, `hue`, `order`, `ci`, `estimator`, `palette`    |
| `sns.boxplot()`     | Diagrama de caja para mostrar distribución y outliers.    | `sns.boxplot(data=df, x="day", y="total_bill")`            | `data`, `x`, `y`, `hue`, `order`, `width`, `palette`, `dodge`     |
| `sns.violinplot()`  | Boxplot + KDE: muestra distribución suavizada.            | `sns.violinplot(data=df, x="sex", y="tip")`                | `data`, `x`, `y`, `hue`, `bw`, `cut`, `inner`, `split`, `palette` |
| `sns.histplot()`    | Histograma univariado o bivariado.                        | `sns.histplot(data=df, x="total_bill", kde=True)`          | `data`, `x`, `y`, `bins`, `kde`, `hue`, `stat`, `element`         |
| `sns.heatmap()`     | Mapa de calor para mostrar valores en forma matricial.    | `sns.heatmap(df.corr(), annot=True)`                      | `data`, `annot`, `cmap`, `linewidths`, `linecolor`, `fmt`         |
| `sns.pairplot()`    | Matriz de gráficos entre variables numéricas.             | `sns.pairplot(df)`                                         | `data`, `hue`, `palette`, `kind`, `diag_kind`, `corner`           |
| `sns.catplot()`     | Constructor general para gráficos categóricos.            | `sns.catplot(data=df, x="day", y="tip", kind="bar")`       | `data`, `x`, `y`, `hue`, `col`, `row`, `kind`, `order`, `height`  |


##  Parámetros comunes en gráficos de Seaborn – Descripción detallada

| Parámetro       | Descripción                                                                                      |
|------------------|--------------------------------------------------------------------------------------------------|
| `data`           | DataFrame que contiene las variables a graficar. Todas las funciones trabajan con `data`.       |
| `x`, `y`         | Nombres de las columnas del DataFrame que se asignan al eje X e Y respectivamente.              |
| `hue`            | Variable categórica o binaria que agrupa los datos por color. Muy útil para comparar subgrupos. |
| `style`          | Cambia el estilo (marcador, línea) en función de una variable categórica.                      |
| `size`           | Cambia el tamaño de los puntos según una variable numérica o categórica.                        |
| `palette`        | Paleta de colores a utilizar (`"pastel"`, `"deep"`, `"Set2"`, `"coolwarm"`, etc.).              |
| `order`          | Lista que define el orden personalizado de las categorías en el eje X.                          |
| `hue_order`      | Orden personalizado para los grupos definidos por `hue`.                                        |
| `kind`           | Tipo de gráfico a generar (en funciones como `catplot`, `relplot`). Ej: `"bar"`, `"box"`, etc. |
| `ax`             | Objeto `Axes` de matplotlib donde se desea insertar el gráfico.                                 |
| `ci`             | Nivel de confianza para las bandas de error. `ci=95` por defecto; usar `None` para desactivarlas.|
| `estimator`      | Función de agregación como `np.mean`, `np.median`, etc. Usada en `barplot`, `pointplot`, etc.   |
| `dodge`          | En gráficos con `hue`, separa las categorías de forma horizontal.                              |
| `linewidth`      | Grosor de líneas o bordes.                                                                       |
| `alpha`          | Transparencia de los elementos (0: invisible, 1: opaco).                                        |
| `kde`            | Si se desea incluir el gráfico de densidad (en histogramas).                                    |
| `annot`          | Muestra los valores numéricos dentro de las celdas (en `heatmap`).                              |
| `bins`           | Cantidad de contenedores (barras) en los histogramas.                                           |
| `diag_kind`      | Tipo de gráfico en la diagonal de un `pairplot`: `"hist"` o `"kde"`.                           |
| `corner`         | Si se omite la parte superior del `pairplot` (cuando es redundante).                            |

```python
# prompt: realizar un grafico para cada tipo de grafico nombrado en la tabla, utilizando la mayoria de parametros opcionales

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker",
    size="size",
    sizes=(50, 200),
    palette="viridis",
    alpha=0.7
)
plt.title("Relación entre total y propina por género, fumador y tamaño de grupo", fontsize=14)
plt.xlabel("Total de la cuenta ($)", fontsize=12)
plt.ylabel("Propina ($)", fontsize=12)
plt.legend(title="Características", loc="upper left")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Gráfico de líneas (ejemplo simple con datos sintéticos)
# Creamos datos de ejemplo que se presten a un lineplot
data_line = pd.DataFrame({
    'tiempo': np.arange(0, 10, 0.5),
    'serie_A': np.sin(np.arange(0, 10, 0.5)) * 10 + np.random.randn(20) * 2,
    'serie_B': np.cos(np.arange(0, 10, 0.5)) * 10 + np.random.randn(20) * 2,
    'categoria': ['Grupo 1'] * 10 + ['Grupo 2'] * 10
})

plt.figure(figsize=(10, 6))
sns.lineplot(
    data=data_line,
    x="tiempo",
    y="serie_A",
    hue="categoria",
    style="categoria",
    markers=True,
    dashes=True,
    ci=95, # Default, included for completeness
    palette="plasma"
)
plt.title("Evolución de Serie A por Categoría a lo largo del Tiempo", fontsize=14)
plt.xlabel("Tiempo (unidades)", fontsize=12)
plt.ylabel("Valor de Serie A", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# Gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(
    data=df,
    x="day",
    y="total_bill",
    hue="time",
    order=["Thur", "Fri", "Sat", "Sun"],
    hue_order=["Lunch", "Dinner"],
    ci=95, # Muestra barras de error
    estimator=np.mean, # Calcula la media
    palette="rocket",
    dodge=True # Separa las barras por 'hue'
)
plt.title("Total de la cuenta promedio por día y horario", fontsize=14)
plt.xlabel("Día de la semana", fontsize=12)
plt.ylabel("Total de la cuenta promedio ($)", fontsize=12)
plt.legend(title="Horario")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()


# Diagrama de caja
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df,
    x="day",
    y="tip",
    hue="smoker",
    order=["Thur", "Fri", "Sat", "Sun"],
    hue_order=["Yes", "No"],
    width=0.6,
    palette="tab10",
    dodge=True # Separa las cajas por 'hue'
)
plt.title("Distribución de propinas por día y estado de fumador", fontsize=14)
plt.xlabel("Día de la semana", fontsize=12)
plt.ylabel("Propina ($)", fontsize=12)
plt.legend(title="Fumador")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(
    data=df,
    x="day",
    y="total_bill",
    hue="sex",
    order=["Thur", "Fri", "Sat", "Sun"],
    hue_order=["Male", "Female"],
    bw='scott', # Estimador de ancho de banda
    cut=2, # Extiende el plot más allá de los datos
    inner="quartile", # Muestra cuartiles en el interior
    split=True, # Parte los violines por 'hue'
    palette="viridis",
    linewidth=1
)
plt.title("Distribución del total de la cuenta por día y género (Violin Plot)", fontsize=14)
plt.xlabel("Día de la semana", fontsize=12)
plt.ylabel("Total de la cuenta ($)", fontsize=12)
plt.legend(title="Género")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

# Histograma
plt.figure(figsize=(10, 6))
sns.histplot(
    data=df,
    x="total_bill",
    hue="time",
    bins=15, # Número de barras
    kde=True, # Superpone la curva de densidad
    stat="density", # Normaliza las alturas para que el área sume 1
    element="step", # Dibuja el histograma como pasos
    palette="coolwarm"
)
plt.title("Distribución del Total de la Cuenta por Horario", fontsize=14)
plt.xlabel("Total de la cuenta ($)", fontsize=12)
plt.ylabel("Densidad", fontsize=12)
plt.legend(title="Horario", labels=["Dinner", "Lunch"]) # Seaborn a veces requiere ayuda con la leyenda de hue en histplot
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()


# Mapa de calor (ejemplo con correlación)
plt.figure(figsize=(8, 6))
correlation_matrix = df.corr(numeric_only=True) # Calcular la matriz de correlación
sns.heatmap(
    correlation_matrix,
    annot=True, # Muestra los valores numéricos en las celdas
    cmap="YlGnBu", # Paleta de color
    linewidths=.5, # Espacio entre celdas
    linecolor='black',
    fmt=".2f" # Formato de los números flotantes
)
plt.title("Matriz de Correlación de Variables Numéricas", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout() # Ajusta el diseño para evitar cortar etiquetas
plt.show()

# Pair plot
# Seleccionamos solo las columnas numéricas y una categórica para el pairplot
df_pair = df[['total_bill', 'tip', 'size', 'sex']]
sns.pairplot(
    data=df_pair,
    hue="sex", # Colorea los puntos por género
    palette="viridis",
    kind="reg", # Superpone una línea de regresión lineal
    diag_kind="kde", # Gráfico de densidad en la diagonal
    corner=True # Muestra solo la parte inferior izquierda (evita redundancia)
)
plt.suptitle("Matriz de Gráficos de Dispersión y Distribución por Género", y=1.02, fontsize=16) # Título general
plt.show()

# Cat plot (ejemplo con barplot usando catplot)
g = sns.catplot(
    data=df,
    x="day",
    y="total_bill",
    hue="smoker",
    col="time", # Crea columnas separadas por 'time'
    kind="bar", # Especifica el tipo de gráfico categórico
    order=["Thur", "Fri", "Sat", "Sun"],
    hue_order=["Yes", "No"],
    ci=None, # Desactiva las barras de error para simplificar
    palette="magma",
    height=5, # Altura de cada facet
    aspect=.8 # Relación de aspecto de cada facet
)
g.fig.suptitle("Total de la cuenta promedio por Día, Fumador y Horario", y=1.03, fontsize=16) # Título general
g.set_axis_labels("Día de la semana", "Total de la cuenta promedio ($)")
g.set_titles("Horario: {col_name}")
g.add_legend(title="Fumador")
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajusta el layout para dar espacio al título principal
plt.show()
```
