# Ciclo FOR

## Ciclo básico

```python
precios = np.random.normal(loc=10, scale=1, size=5) # Normal de media 10 sigma 1
precios
```

```python
# Una lista por comprension es en realidad un ciclo iterativo FOR abreviado

precios_nuevos = [precio + 5 for precio in precios]

precios_nuevos
```

```python
precios_nuevos = []

# for x->iterable en COLECCION :
for x in precios:
  precios_nuevos.append(round(x+5, 2))
  print(f'Estoy pasando por el bloque con el numero {x:.2f}')

print("Ya termine, ahi va el resuyltado final", precios_nuevos)
```

```python
import time

precios_nuevos = []

for x in precios:
  precios_nuevos.append(round(x+5, 2))
  print(f'Estoy demorando 3 segundos para procesar el numero {x:.2f}')
  time.sleep(3)


print("Ya termine, ahi va el resuyltado final", precios_nuevos)
```

##  Generar la lista de retornos

Tengan en cuenta que si bien es lo mismo usar una iteracion y el paquete random nativo que usar numpy, el hecho de generar en forma vectorial/matricial todo a partir de la misma seed en un solo paso es infinitamente mas eficiente

```python
import random
```

```python
%%timeit

# Usando el randomizador de numpy
n = 5000
ret = np.random.normal(0,2,n)
```

```python
%%timeit

# Usando el randomizador nativode python
n = 5000
ret = [ random.normalvariate(mu=0, sigma=2)  for i in range(n)]
```

```python
n = 100_000
ret = np.random.normal(0,2,n)
ret
```

```python
ret.std()
```

image.png

```python
mu = ret.mean()

sigma = 0
for xi in ret:
  sigma += (xi - mu)**2

sigma = (sigma/(n-1))**0.5
round(sigma,15)
```

```python
%%timeit
ret.std()
```

```python
%%timeit
mu = ret.mean()

sigma = 0
for xi in ret:
  sigma += (xi - mu)**2

sigma = (sigma/(n-1))**0.5
round(sigma,15)
```
